```python
from matplotlib import pyplot as plt
import matplotlib.image as img
import matplotlib
import numpy as np
import pandas as pd
import csv
```

```python
df = pd.read_csv('data.csv')
SNo = df["S.No"]
Designation = df["Designation"]
Grade = df["Grade"]
Location = df["Location"]
Gender = df["Gender"]
Education = df["Education"]
Age = df["Age"]
Lr = df["Last Rating"]
Mi = df["Monthly Income"]
Engagement = df["Engagement Score (% Satisfaction)"]
Ms = df["Marital Status"]
Zone = df["Zone"]
Remarks = df["Remarks"]
```

```python
plt.figure()
Grade.value_counts().plot(kind='bar', xlabel='Grades', ylabel='frequency', figsize=(10,5))

plt.show()
```

Majority of the people who left had E1 and E2 grades.

```python
Tenure = df["Tenure"]
for i in range(np.size(Tenure)):
  Tenure[i] = float(Tenure[i].replace(" ", ""))

```

```python
plt.figure()
Education.value_counts().plot(kind='bar', xlabel='Education', ylabel='frequency', figsize=(10,5))

plt.show()
```

Very high percentage of the employees were bachelors in educational qualification

```python
plt.figure()
Designation.value_counts().plot(kind='barh', xlabel='Designation', ylabel='frequency', figsize=(10,5))

plt.show()
```

```python
plt.scatter(Tenure, Designation)
```

Most of the employees had low tenures( < 4 years) and most of them were sales executives

```python
plt.figure()
Tenure.value_counts().plot(kind='bar', xlabel='Tenure', ylabel='frequency', figsize=(15,5))

plt.show()
```

Most of the employees had tenures less than 1 year

```python
plt.figure()
Lr.value_counts().plot(kind='pie', xlabel='Last Rating', figsize=(10,5))

plt.show()
```

```python
plt.figure()
Age.value_counts().plot(kind='bar', xlabel='Age', figsize=(10,5))

plt.show()
```

Employees between the ages 24-30 were the most probable to quit

```python
for i in range(Engagement.size):
  Engagement[i] = float(Engagement[i].replace("%", ""))
```

```python
plt.figure()
Engagement.value_counts().plot(kind='bar', xlabel='Engagement', ylabel='frequency', figsize=(15,8))

plt.show()
```

Engagement shows no specific dependence.

```python
plt.scatter(Age, Tenure)
```

```python
plt.figure()
for i in range(Zone.size):
  Zone[i] = Zone[i].upper()
Zone.value_counts().plot(kind='barh', xlabel='Last Rating', figsize=(10,5))

plt.show()
```

People living living in the south zone quit most. Work location would be in the North zone it seems.

```python
plt.figure()
Ms.value_counts().plot(kind='bar', xlabel='Marital Status', ylabel='Frequency', figsize=(10,5))

plt.show()
```
